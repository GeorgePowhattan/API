```python
# Make a dataframe with date and value as columns
```

```python
dates = [item['date'] for item in parsed['observations'] ]
```

```python
values = [item['value'] for item in parsed['observations'] ]
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame(data=values, index=dates, dtype=float)
```

```python
df = df.rename(columns={0: "GDP"})
```
